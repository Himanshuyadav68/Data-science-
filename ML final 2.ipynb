```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report, confusion_matrix
from imblearn.under_sampling import NearMiss
# Load the CSV data
data = pd.read_csv('data.csv')
print(data.head())
```

```python
import numpy as np

data['abs_diff'] = np.abs(np.abs(data['newbalanceOrig'] - data['oldbalanceOrg'])-data['amount'])

# Filter rows based on  condition (absolute difference <= 2)

data = data[(data['abs_diff'] <= 2) | (data['isFlaggedFraud'] == 1)]

# Droping the 'abs_diff' column as we  don't need it in the final output
data= data.drop(columns=['abs_diff'])
print("after cleaning")
print(len(data))
```

```python
def categorize_nameDest(name):
  if name[0] == "C":
    return 1
  elif name[0] == "M":
    return 0
  else:
    raise ValueError(f"Unknown nameDest category: {name}")

data["nameDest"] = data["nameDest"].apply(categorize_nameDest)
data_processed=data.drop(["nameOrig", "oldbalanceDest", "newbalanceDest","type"], axis=1)
print(data_processed.head())
# One-hot encode categorical data
data_encoded=data_processed
print(data_encoded.head())
```

played in bar plots.

```python
from sklearn.metrics import precision_recall_curve, f1_score, auc
from imblearn.over_sampling import ADASYN


# Separate features and target variables
X = data_encoded.drop(["isFraud", "isFlaggedFraud"], axis=1)
y_fraud = data_encoded["isFraud"]
y_flagged = data_encoded["isFlaggedFraud"]
# Replace NearMiss with ADASYN:
adasyn = ADASYN(random_state=42)  # Optionally adjust other ADASYN parameters
X_res_fraud, y_res_fraud = adasyn.fit_resample(X, y_fraud)

# Split data into training and testing sets (assuming you have X_test and y_test)
X_train_fraud, X_test_fraud, y_train_fraud, y_test_fraud = train_test_split(X_res_fraud, y_res_fraud, test_size=0.2, random_state=42)

# Train Random Forest models on the undersampled data
model_fraud = RandomForestClassifier(random_state=42)
model_fraud.fit(X_train_fraud, y_train_fraud)  # Use training sets for fitting
```

```python
#____________________________________________
from sklearn.metrics import precision_recall_curve, f1_score, auc
from imblearn.over_sampling import SMOTE

# Apply SMOTE oversampling for isFlaggedFraud
sm = SMOTE(random_state=42)
X_res_flagged, y_res_flagged = sm.fit_resample(X, y_flagged)


# Combine flagged and normal transactions
combined_data = pd.concat([X_res_flagged[y_res_flagged == True], X_res_flagged[y_res_flagged == False]])
combined_target = pd.concat([y_res_flagged[y_res_flagged == True], y_res_flagged[y_res_flagged == False]])

# Split combined data into training and testing sets with a 40/60 split
X_train_flagged, X_test_flagged, y_train_flagged, y_test_flagged = train_test_split(
    combined_data, combined_target, test_size=0.4, random_state=42
)

# Train model on the combined data
model_flagged = RandomForestClassifier(random_state=42)
model_flagged.fit(X_train_flagged, y_train_flagged)

```

```python
# Evaluate fraud model
y_pred_fraud = model_fraud.predict(X_test_fraud)  # Use testing set for prediction
precision_fraud, recall_fraud, thresholds_fraud = precision_recall_curve(y_test_fraud, y_pred_fraud)
f1_fraud = f1_score(y_test_fraud, y_pred_fraud)
pr_auc_fraud = auc(recall_fraud, precision_fraud)


print("Fraud Detection Model:")
print(f"Precision-Recall AUC: {pr_auc_fraud:.4f}")
print(f"F1-score: {f1_fraud:.4f}")

# Evaluate flagged fraud model
y_pred_flagged = model_flagged.predict(X_test_flagged)  # Use testing set for prediction
precision_flagged, recall_flagged, thresholds_flagged = precision_recall_curve(y_test_flagged, y_pred_flagged)
f1_flagged = f1_score(y_test_flagged, y_pred_flagged)
pr_auc_flagged = auc(recall_flagged, precision_flagged)

print("\nFlagged Fraud Model:")
print(f"Precision-Recall AUC: {pr_auc_fraud:.4f}")
print(f"F1-score: {f1_fraud:.4f}")
import matplotlib.pyplot as plt
#  visualize precision-recall curves
plt.plot(recall_fraud, precision_fraud, label="Fraud Model")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curves")
plt.legend()
plt.show()
```

```python
plt.plot(recall_flagged, precision_flagged, label="Flagged Fraud Model")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curves")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
# Evaluate the model for 'isFraud'
print("Confusion Matrix for isFraud:\n", confusion_matrix(y_test_fraud, y_pred_fraud))
print("\nClassification Report for isFraud:\n", classification_report(y_test_fraud, y_pred_fraud))

# Evaluate the model for 'isFlaggedFraud'
print("\nConfusion Matrix for isFlaggedFraud:\n", confusion_matrix(y_test_flagged, y_pred_flagged))
print("\nClassification Report for isFlaggedFraud:\n", classification_report(y_test_flagged, y_pred_flagged))
```
